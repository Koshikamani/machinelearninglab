```python
#19  prepare x and y
x=df.drop(columns=['Id','Species'])
y=df['Species']
#print(x)
#print(y)
```

```python
#20 split into train and test data 70% train random state =1
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=1)
#print(x_train)
```

```python
#21 look at documentation of svc class then use it to create a  model
from sklearn.svm import SVC
help(SVC)
model=SVC()
```

```python
#22 Model buiding
model.fit(x_train,y_train)
```

```python
#23 instatiate an object and CSVSC using cost c=1
svc_model=SVC(C=1)
```

```python
#24 fit the model
svc_model.fit(x_train,y_train)
```

```python
#25predict the model 
y_pred=svc_model.predict(x_test)
print(y_pred[:5])
```

```python
#26 evaluate confusion matrix
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,classification_report
confusion_matrix(y_test,y_pred)
```

```python
#27 print other metrics
print("accuracy:",accuracy_score(y_test,y_pred))
print("classification report:",classification_report(y_test,y_pred))
```
